```python
# Deletando as colunas desnecessárias
# axis=1 indica que estamos removendo colunas (e não linhas)
# inplace=True aplica a alteração diretamente no dataframe atual
dataframe.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'Cabin'], axis=1, inplace=True)
```

```python
# contando a quantidade de valores nulos no dataframe

dataframe.isnull().sum()
```

```python
dataframe['Embarked'] = dataframe['Embarked'].fillna('S')
```

```python
# Verificando o valor médio da coluna 'Age'
dataframe['Age'].mean()
```

```python
dataframe['Age'] = dataframe['Age'].fillna(dataframe['Age'].mean())
```

```python
# Verificando se ainda resta algum valor nulo no dataframe
dataframe.isnull().sum()
```

```python
dataset = dataframe
```

```python
dataset
```

pré-processamento

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
```

```python
fare = dataset['Fare'].values
```

```python
fare
```

```python
fare_scaled = scaler.fit_transform(fare.reshape(-1, 1))
```

```python
fare_scaled
```

```python
dataset['Fare'] = fare_scaled
```

```python
dataset.head()
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
fare_scaled = scaler.fit_transform(fare.reshape(-1, 1))
```

```python
dataset['Fare'] = fare_scaled
```

```python
dataset.head()
```

```python
encoder = preprocessing.OrdinalEncoder()
```

```python
sex = dataset['Sex'].values
```

```python
sex_encoder = encoder.fit_transform(sex.reshape(-1, 1))
```

```python
dataset['Sex_encoded'] = sex_encoder
```

```python
dataset.head()
```

```python
one_hot_encoder = preprocessing.OneHotEncoder()
```

```python
sex_one_hot_encoder = one_hot_encoder.fit_transform(dataset['Sex'].values.reshape(-1, 1))
```

```python
one_hot_encoder.get_feature_names_out(['Sex'])
```

```python
sex_one_hot_encoder_values = sex_one_hot_encoder.toarray()
```

```python
sex_one_hot_encoder_values.shape
```

```python
dataset[one_hot_encoder.get_feature_names_out(['Sex'])] = sex_one_hot_encoder_values
```
